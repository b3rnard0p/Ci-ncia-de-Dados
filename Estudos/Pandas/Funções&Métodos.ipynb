```python
import pandas as pd

caminho = '/content/drive/MyDrive/Ciência de Dados/Dados/Vendas.csv'
arquivo = pd.read_csv(caminho)
```

```python
# EXERCÍCIO 1
# Variação A:
print(arquivo.tail())
print(arquivo.columns)

# Variação B:
print(arquivo.tail(3))
lista_colunas = list(arquivo.columns)
print("Lista de colunas:", lista_colunas)
```

```python
# EXERCÍCIO 2
# Variação A:
print(pd.crosstab(index=arquivo['Produto'], columns='Contagem'))

# Variação B:
contagem_produtos = arquivo['Produto'].value_counts()
print(contagem_produtos)
```

```python
# EXERCÍCIO 3
# Variação A:
print(arquivo.loc[arquivo['Faturamento'] > 500])

# Variação B:
vendas_altas = arquivo.loc[arquivo['Faturamento'] > 500]
print(vendas_altas)
```

```python
# EXERCÍCIO 4
# Variação A:
linha_nula = arquivo.loc[arquivo['Faturamento'].isna()]
print("Linha nula encontrada:\n", linha_nula)
arquivo['Faturamento'] = arquivo['Faturamento'].fillna(0)

# Variação B:
id_afetado = arquivo.loc[arquivo['Faturamento'].isna(), 'ID_Venda']
print("ID afetado:\n", id_afetado)
arquivo['Faturamento'] = arquivo['Faturamento'].fillna(0)
```

```python
# EXERCÍCIO 5
# Variação A:
print(arquivo.groupby('Produto')['Quantidade'].mean())

# Variação B:
media_quantidade = arquivo.groupby('Produto').agg({'Quantidade': 'mean'})
print(media_quantidade)
```

```python
# EXERCÍCIO 6
# Variação A:
print(arquivo.groupby('Data')['Faturamento'].sum())

# Variação B:
faturamento_dia= arquivo.groupby('Data').agg({'Faturamento': 'sum'})
print(faturamento_dia)
```

```python
# EXERCÍCIO 7
# Variação A:
print(arquivo.iloc[4:8, [1, 2]])

# Variação B:
print(arquivo.iloc[4:8, 1:3])
```

```python
# EXERCÍCIO 8
# Variação A:
arquivo['Ticket_Medio'] = arquivo['Faturamento'] / arquivo['Quantidade']
print(arquivo.head(2))

# Variação B:
arquivo = arquivo.assign(Ticket_Medio = arquivo['Faturamento'] / arquivo['Quantidade'])
print(arquivo.head(2))
```

```python
# EXERCÍCIO 9
# Variação A:
arquivo['Data'] = pd.to_datetime(arquivo['Data'])
print(arquivo.loc[arquivo['Data'] > '2024-01-02'])

# Variação B:
arquivo['Data'] = pd.to_datetime(arquivo['Data'])
filtro_booleano = arquivo['Data'] > '2024-01-02'
print(arquivo.loc[filtro_booleano])
```

```python
# EXERCÍCIO 10
# Variação A:
agrupado_a = arquivo.groupby('Produto')[['Quantidade', 'Faturamento']].sum()
filtrado_a = agrupado_a.loc[agrupado_a['Quantidade'] > 3]
ranking_a = filtrado_a.sort_values(by='Faturamento', ascending=False)
print(ranking_a)

# Variação B:
agrupado_b = arquivo.groupby('Produto')[['Quantidade', 'Faturamento']].sum()
filtrado_b = agrupado_b.loc[agrupado_b['Quantidade'] > 3]
print(filtrado_b.sort_values(by='Faturamento', ascending=False))
```
